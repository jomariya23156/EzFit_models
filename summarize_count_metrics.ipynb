```python
import os
import numpy as np
import pandas as pd
```

## UCF_dataset

```python
BENCHMARK = True
POST = 'None'
csv_dir = '/data1-6tb/jom/EzFit_models/evaluation_videos/UCF_dataset/outputs/'
```

```python
# # model_types = ['kps_stacked', 'kps_seq', 'kps_single', 'image_single', 'ViViT', 'video_sequence']
# model_types = ['kps_seq']
# exercise_list = ['pushup', 'squat', 'jumping-jack']
# for model in model_types:
#     temp = os.path.join(csv_dir, f'{model}_hard_vote_UCF_dataset.csv')
#     print(f'===== {model} =====')
#     df = pd.read_csv(temp)
#     print('Overall OBO:', df['obo'].mean())
#     for exercise in exercise_list:
#         print(f'OBO of {exercise}:', df[df['exercise']==exercise]['obo'].mean())
#     print()

# model_types = ['kps_stacked', 'kps_seq', 'kps_single', 'image_single', 'ViViT', 'video_sequence']
model_types = ['kps_single']
exercise_list = ['pushup', 'squat', 'jumping-jack']
for model in model_types:
    temp = os.path.join(csv_dir, f'{model}_{POST}_UCF_dataset.csv')
    if BENCHMARK:
        temp = os.path.join(csv_dir, f'{model}_{POST}_UCF_dataset_BM.csv')
    print(f'===== {model} =====')
    print('==================')
    df = pd.read_csv(temp)
    # not include jumping-jack_extra (full side view)
    print('Overall OBO:', df[~df['file_name'].str.contains('jumping-jack_extra')]['obo'].mean())
    for exercise in exercise_list:
        print(f'----->{exercise}<-----')
        if exercise == 'jumping-jack':
            res_df = df[df['exercise']==exercise][~df['file_name'].str.contains('extra')]
            res_df['5-classes-view'] = res_df['5-classes-view'].str.strip()
            print(f'OBO of {exercise}:', res_df['obo'].mean())
        else:
            res_df = df[df['exercise']==exercise]
            print(f'OBO of {exercise}:', res_df['obo'].mean())
        
        ### labels analysis ###
        # view
        for view in res_df['5-classes-view'].unique():
            temp = res_df[res_df['5-classes-view']==view]
            print(f'===View: {view}===')
            print(f'amount: {len(temp)} | percentage: {len(temp)/len(res_df)}')
            print(f"OBO: {temp['obo'].mean()}")
            print()
        # position
        temp = res_df[res_df['position']=='center']
        print(f'===Position: center===')
        print(f'amount: {len(temp)} | percentage: {len(temp)/len(res_df)}')
        print(f"OBO: {temp['obo'].mean()}")
        print()
        temp = res_df[res_df['position']!='center']
        print(f'===Position: Non-center===')
        print(f'amount: {len(temp)} | percentage: {len(temp)/len(res_df)}')
        print(f"OBO: {temp['obo'].mean()}")
        print()
        # distance
        for dist in res_df['distance'].unique():
            temp = res_df[res_df['distance']==dist]
            print(f'===Distance: {dist}===')
            print(f'amount: {len(temp)} | percentage: {len(temp)/len(res_df)}')
            print(f"OBO: {temp['obo'].mean()}")
            print()
        # gender
        for gender in res_df['gender'].unique():
            temp = res_df[res_df['gender']==gender]
            print(f'===Gender: {gender}===')
            print(f'amount: {len(temp)} | percentage: {len(temp)/len(res_df)}')
            print(f"OBO: {temp['obo'].mean()}")
            print()
        # indoor/outdoor
        for area in res_df['indoor/outdoor'].unique():
            temp = res_df[res_df['indoor/outdoor']==area]
            print(f'===Indoor/Outdoor: {area}===')
            print(f'amount: {len(temp)} | percentage: {len(temp)/len(res_df)}')
            print(f"OBO: {temp['obo'].mean()}")
            print()
        
        print('-------------------')
    print()
```
