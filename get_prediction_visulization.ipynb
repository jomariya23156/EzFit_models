```python
from collections import Counter
import cv2
import pickle
import numpy as np
```

```python
# 0 = not related
# 1 = squat-down
# 2 = squat-up
vid_num = 3
ground_truth = []
vid_path = f'objective_test_videos/squat/squat_{vid_num}.mp4'
cap = cv2.VideoCapture(vid_path)
total_frame = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
while cap.isOpened():
    grab, frame = cap.read()
    if not grab:
        break
    
    cv2.imshow('video', cv2.resize(frame, (640, 368)))
    # cv2.waitKey(0) return ascii and chr() convert ascii to string
    key = chr(cv2.waitKey(0))
    print(key, end='\r')
    try:
        key = int(key)
    except:
        print('[ERR] Invalid label: input ->', key)
        break
    ground_truth.append(key)
    
cap.release()
cv2.destroyAllWindows()
assert len(ground_truth) == total_frame, "# of label != # of frames in the video"
```

```python
with open(f'saved_pickles/for_visualization/squat/test_vid_{vid_num}/test_vid_{vid_num}_gt.pickle', 'wb') as f: 
    pickle.dump(ground_truth, f)
```
